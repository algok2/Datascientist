```python
def stock_price(ticker):
    url = f'https://finance.naver.com/item/main.naver?code={ticker}'
    res = requests.get(url = url)
    soup = BTS(markup = res.text, features = 'html.parser')
    items = soup.select(selector = 'div#rate_info_krx > div > p.no_today > em > span')
    return items[0].text
```

```python
stock_price(ticker = '005930')
```

```python
stock_price(ticker = '000660')
```

```python
tickers = ['005930', '000660', '402340', '005380', '373220']
```

```python
import time
```

```python
for ticker in tickers:
    price = stock_price(ticker)
    print(price)
    time.sleep(1)
```

```python
compNms = ['삼성전자', '하이닉스', 'SK스퀘어', '현대차', '엘지엔솔']
```

```python
# zip 함수는 반복 가능한 객체의 같은 인덱스 원소를 튜플로 묶어서 반환합니다.
for i in zip(tickers, compNms):
    print(i)
```

```python
for i, j in zip(tickers, compNms):
    print(j)
```

```python
for ticker, compNm in zip(tickers, compNms):
    price = stock_price(ticker)
    print(f'{compNm}의 현재 가격은 {price}원입니다.')
    time.sleep(1)
```
