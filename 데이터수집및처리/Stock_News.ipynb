```python
import requests
```

```python
url = 'https://finance.naver.com/news/mainnews.naver'
```

```python
# HTTP 요청이 실행되면서 쿼리 문자열의 정수가 자동으로 문자열로 변환됩니다.
query = {
    'date': '2026-01-01',
    'page': 1
}
```

```python
res = requests.get(url = url, params = query)
```

```python
class MyBTS(BTS):
    def html_text(self, css):
        items = self.select(css)
        return [item.text for item in items]
    def html_attr(self, css, key = 'href'):
        items = self.select(css)
        return [item[key] for item in items]
```

```python
soup2 = MyBTS(markup = res.text, features = 'html.parser')
```

```python
# 제목
soup2.html_text(css = '.articleSubject > a')
```

```python
# 링크
soup2.html_attr(css = '.articleSubject > a')
```

```python
# 언론사
soup2.html_text(css = 'span.press')
```

```python
# 등록일시
soup2.html_text(css = 'span.wdate')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data = {
    '제목': soup2.html_text(css = '.articleSubject > a'),
    '링크': soup2.html_attr(css = '.articleSubject > a'), 
    '언론사': soup2.html_text(css = 'span.press'), 
    '등록일': soup2.html_text(css = 'span.wdate')
})
```

```python
df['링크']
```

```python
# 후방참조: 'article_id=' 다음에 오는 숫자를 추출
df['AID'] = df['링크'].str.extract(pat = '((?<=article_id=)[0-9]+)')
```

```python
# 후방참조: 'office_id=' 다음에 오는 숫자를 추출
df['OID'] = df['링크'].str.extract(pat = '((?<=office_id=)[0-9]+)')
```
