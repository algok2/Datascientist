```python
import requests
```

```python
url = 'https://finance.naver.com/marketindex/'
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
res.headers['content-type']
```

```python
from bs4 import BeautifulSoup as BTS
```

```python
soup = BTS(markup = res.text, features = 'html.parser')
```

```python
items = soup.select('#exchangeList > li > a.head > h3 > span')
len(items)
```

```python
[item.text for item in items]
```

```python
items = soup.select('#exchangeList > li > a.head > div > span.value')
len(items)
```

```python
[item.text for item in items]
```

```python
class MyBTS(BTS):
    def html_text(self, css):
        items = soup.select(css)
        return [item.text for item in items]
```

```python
soup2 = MyBTS(markup = res.text, features = 'html.parser')
```

```python
# dir 함수는 변수가 가리키는 객체의 속성과 메서드 목록을 반환합니다.
dir(soup)
```

```python
soup2.html_text(css = '#exchangeList > li > a.head > h3 > span')
```

```python
soup2.html_text(css = '#exchangeList > li > a.head > div > span.value')
```
