```python
apt['거래금액'] = apt['거래금액']/10000
apt.head()
```

```python
apt['경과년수'] = apt['계약년']- apt['입주년도']
apt.head()
```

```python
apt.rename(columns = {'계약년':'계약년도'}, inplace=True)
apt.head()
```

```python
cond = apt['경과년수'].ge(30)
apt['재건축'] = np.where(cond, '충족', '부족')
apt.head()
```

```python
apt.dtypes
```

```python
apt['계약년도'] = apt['계약년도'].astype(str)
apt['계약월'] = apt['계약월'].astype(str)
apt['계약일'] = apt['계약일'].astype(str)
```

```python
apt.dtypes
```

```python
apt = apt.astype({'계약년도':'int64', '계약월':'int64', '계약일':'int64'})
apt.dtypes
```

```python
cols = ['계약년도', '계약월', '계약일']
apt[cols] = apt[cols].astype(str)
apt.dtypes
```

```python
apt['계약일자'] = apt['계약년도']+'-'+apt['계약월']+'-'+apt['계약일']
apt['계약일자'] = apt['계약일자'].astype('datetime64[ns]')
```
