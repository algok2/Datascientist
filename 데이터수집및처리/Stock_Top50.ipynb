```python
import requests
```

```python
url = 'https://finance.naver.com/sise/sise_market_sum.naver?&page=1'
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
res.headers['content-type']
```

```python
from bs4 import BeautifulSoup as BTS
```

```python
soup = BTS(markup = res.text, features = 'html.parser')
```

```python
items = soup.select('table.type_2 > tbody > tr > td.number:nth-child(3)')
len(items)
```

```python
items[0].text
```

```python
# 시가총액 상위 50위 종목의 현재가를 출력합니다.
for item in items:
    print(item.text)
```

```python
# 시가총액 상위 50위 종목의 현재가를 리스트 원소로 저장합니다.
result = []
for item in items:
    result.append(item.text)
```

```python
result
```

```python
# 리스트 컴프리헨션(내포)
[item.text for item in items]
```
