```python
tag_n = ["NNG", "NNP"]
tag_v = ['VV', 'VA', 'VV-R', 'VA-I','VV-R', 'VA-I']
```

```python
[token.form for token in tokens if token.tag in tag_n+tag_v ]
```

```python
[token.form +'다' if token.tag in tag_v else token.form for token in tokens if token.tag in tag_n+tag_v ]
```

```python
def tokenizer(text):
    tokens = kiwi.tokenize(text, stopwords=stopwords)
    return [token.form +'다' if token.tag in tag_v else token.form for token in tokens if token.tag in tag_n+tag_v ]
```

```python
tokenizer(text=corrected)
```

```python
from tqdm.notebook import tqdm
```

```python
morphs = []

for corrected in tqdm(df['corrected']):
    tokens = tokenizer(text = corrected)
    morphs.append(tokens)
```

```python
words = ['사무용', '저렴', '재구매', '구매 확정']
tags = ['NNG']*5
```

```python
for word, tag in zip(words, tags):
    kiwi.add_user_word(word, tag)
```

```python
morphs = []

for corrected in tqdm(df['corrected']):
    tokens = tokenizer(text = corrected)
    morphs.append(tokens)
```

```python
morphs = [[j.replace(' ','') for j in i] for i in morphs]
```

```python
print(morphs)
```

```python
corpus = [' '.join(i) for i in morphs]
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(min_df = 0.01)
```

```python
dtm = cv.fit_transform(corpus).toarray()
```

```python
dtm
```

```python
dtm = pd.DataFrame(data = dtm, columns=cv.get_feature_names_out())
dtm
```
